```python
df['운동량_cut'] = pd.cut(df['운동량'], bins=10) #bins = 구간,  # 일정한 값을 기준으로 구간을 나눔
df['운동량_cut'].value_counts()
```

```python
df1['운동량_qcut'] = pd.qcut(df1['운동량'], q=10) #비율을 기준으로 구간 나눔
df1['운동량_qcut'].value_counts()
```

```python
qcut_bins = [0, 0.2, 0.8, 1] #하위 20%, 상위 20% 버리고 상위 20%~80% 사이 구간을 집중적으로 볼 때 사용. 93.414: 하위 20% 기준치 /3328.186: 상위 20%(하위80%) 기준치ㅊ
pd.qcut(df1['운동량'], qcut_bins)
```

```python
pd.qcut(df1['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
```
