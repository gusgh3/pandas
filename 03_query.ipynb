### 연습

```python
tips = sns.load_dataset('tips')
tips.head()
```

- 남성 흡연자 조건 필터링
- tip을 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

-식사인원이 5명 이상인 저녁테이블

-흡연자가 없는 주말 테이블

-total_bill,tip 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

```python
c1 = tips['size'] >= 5
c2 = tips['time'] == 'Dinner'

c3 = tips['smoker'] == 'No'
c4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(c1 & c2) & (c3 & c4), ['total_bill', 'tip']].head()
```
