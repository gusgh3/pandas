```python
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

## apply()

인자에 다른 여러 함수를 받는, 함수

```python
df.head()
```

```python
df['who'].value_counts()
```

```python
def trans_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df['who'].apply(trans_who)
```

```python
def fare_age(x): #여기서 x는 하나의 열
    return x['fare'] / x['age']
    
```

```python
df.apply(fare_age, axis=1)
```

```python
df['survived'].value_counts()
```

```python
df['survived'].apply(lambda x: '생존' if x == 1 else '사망')
```

### groupby

내가 데이터를 바라볼 때의 기준을 잡아서 그룹화(묶어두는 것). 콜럼들도 그룹화에 따라감

```python
df.head()
```

```python
df.groupby('sex').mean(numeric_only=True)
```

```python
df.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True) # 위 표에서 survivied 콜럼에 해당하는 데이터들만 추출
```

```python
pd.DataFrame(df.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)) #DataFrame으로 형변환 방법 1
```

```python
df.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True) #DataFrame으로 형변환 방법 2
```

## reset_index

인덱스를 초기화하는 방법

```python
df.groupby(['sex', 'pclass'])['survived'].mean().reset_index() #필요한 값을 한눈에 가져오고 싶을 때 인덱스를 다시 초기화하고 재정립
```

```python
df.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
```

## pivot_table

```python
df.pivot_table(index='who', values='survived')
```

```python
df.pivot_table(columns='who', values='survived')
```

```python
df.pivot_table(index=['sex', 'pclass'], values='survived')
```

```python
df.pivot_table(index=['sex', 'pclass'], values = 'survived', aggfunc=['sum', 'mean'])
```
