### fillna

데이터 넣기?

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
```

```python
df1['age'].fillna(100)
```

```python
df1['age'] = df1['age'].fillna(100) #데이터프레임(df) 함수를 실행할 때는 재할당을 꼭 해줘야 한다. 데이터프레임 함수 자체가 값을 함수에 따라 변하게 해주지 않기 때문
```

```python
df1.tail()
```

```python
df1['deck'].fillna('C')
```

```python
# df['deck'].fillna('No Data')
# 카테고리 타입은, 카테고리 범위에 없는 정보는 넣을 수 없음.
```

```python
df1['deck'] = df1['deck'].cat.add_categories('No Data')
df1['deck'].fillna('No Data')
```

```python
df1 = df.copy()
```

```python
mean_age = df1['age'].mean()
df1['age'].fillna(mean_age).tail()
```

```python
median_age = df1['age'].median()
df1['age'].fillna(median_age).tail()
```

-남자 승객의 결측치 => 남자 승객의 평균으로 채우고
-여자 승객의 결측치 => 여자 승객의 평균으로 채우기

```python
df2 = df.copy()
```

```python
c2 = df2['sex'] == 'male'
c2
```

```python
n2 = df2.loc[c2, 'age'].mean(numeric_only=True)
n2
```

```python
n2 = df2.loc[c2, 'age'].mean(numeric_only=True)
#df2.loc[c2, 'age'] = df2.loc[c2, 'age'].fillna(n2)
df2.loc[c2, 'age'].fillna(n2)
df2.head(10)
```

```python
c3 = df2['sex'] == 'female'
c3
```

```python
n3 = df2.loc[c3, 'age'].mean(numeric_only=True)
n3
```

```python
n3 = df2.loc[c3, 'age'].mean(numeric_only=True)
df2.loc[c3, 'age'] = df2.loc[c3, 'age'].fillna(n3)
```

```python
df2['age'].mean()
```

```python
## 강사 답안
```

```python
c1 = df1['sex'] == 'male'
male_mean = df1.loc[c1, 'age'].mean()
male_mean
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(male_mean)

c2 = df1['sex'] == 'female'
female_mean = df1.loc[c2, 'age'].mean()
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(female_mean)
```

```python
df1['age'].mean()
```
